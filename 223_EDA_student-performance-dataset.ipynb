```python
import pandas as pd

path_file = '/content/drive/MyDrive/Learn_AI_Terapan/student_performance.csv'

# Membaca file CSV
df = pd.read_csv(path_file, sep=';')

print("--- Menampilkan 5 Baris Pertama Data (Head) ---")
display(df.head())
```

```python
print("--- 1. Informasi Struktur Dataset (Info) ---")
# Menampilkan tipe data tiap kolom dan mengecek adanya nilai null (kosong)
display(df.info())
```

```python
print("\n--- 2. Menampilkan 5 Baris Terakhir Data (Tail) ---")
# Menampilkan apakah data terbaca penuh hingga terakhir
display(df.tail())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur ukuran kanvas gambar
plt.figure(figsize=(8, 5))

# Membuat grafik distribusi (Histogram + kurva KDE) untuk Exam Score
sns.histplot(df['ExamScore'], bins=20, kde=True, color='skyblue')

# Menambahkan judul dan label
plt.title('Gambar Distribusi Data: Nilai Ujian (Exam Score)', fontsize=14)
plt.xlabel('Nilai Ujian', fontsize=12)
plt.ylabel('Frekuensi (Jumlah Siswa)', fontsize=12)

# Menampilkan gambar grafik
plt.show()
```

5. Membuat grafik sederhana

```python
# Menghitung rata-rata nilai ujian berdasarkan level stres
rata_rata_nilai = df.groupby('StressLevel')['ExamScore'].mean().reset_index()

plt.figure(figsize=(7, 5))

# Membuat diagram batang dengan warna yang berbeda untuk setiap kategori
plt.bar(rata_rata_nilai['StressLevel'], rata_rata_nilai['ExamScore'], color=['#ff9999', '#66b3ff', '#99ff99'])

plt.title('Grafik Sederhana: Rata-rata Nilai Ujian berdasarkan Level Stress', fontsize=14)
plt.xlabel('Tingkat Stress (Stress Level)', fontsize=12)
plt.ylabel('Rata-rata Nilai Ujian', fontsize=12)

# Menampilkan grafuk
plt.show()
```

```python
print("--- Operasi Olah Data Sederhana ---")

# 1. MIN: Mencari Nilai Terendah
min_age = df['Age'].min()
print(f"1. Usia Terendah (Min): {min_age} Tahun")

# 2. MAX: Mencari nilai tertinggi
max_score = df['ExamScore'].max()
print(f"2. Nilai Ujian Tertinggi (Max): {max_score}")

# 3. JUMLAH: Total keseluruhan
total_study_hours = df['StudyHours'].sum()
print(f"3. Total Seluruh Jam Belajar Siswa (Jumlah): {total_study_hours} Jam")

# 4. KURANG: Pengurangan antar 2 variabel
df['WaktuBebas'] = 40 - df['StudyHours']

print("\n--- 5 Baris Pertama Data dengan Kolom Baru 'WaktuBebas' ---")
# Menampilkan kolom StudyHours, ExamScore, dan WaktuBebas untuk pembuktian
display(df[['StudyHours', 'ExamScore', 'WaktuBebas']].head())
```
